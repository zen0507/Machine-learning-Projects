## DBSCAN

**Since the values of the features are in different orders of magnitude, let's scale the data. Use StandardScaler to scale the data.**

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
scaled_X = scaler.fit_transform(df)
```

```python
scaled_X
```

**Use DBSCAN and a for loop to create a variety of models testing different epsilon values. Set min_samples equal to 2 times the number of features. During the loop, keep track of and log the percentage of points that are outliers. For reference the solutions notebooks uses the following range of epsilon values for testing:**

    np.linspace(0.001,3,50)

```python
from sklearn.cluster import DBSCAN
```

```python
outlier_percent = []

for eps in np.linspace(0.001,3,50):

    dbscan = DBSCAN(eps=eps,min_samples=2*scaled_X.shape[1])
    dbscan.fit(scaled_X)

    perc_outliers = 100*np.sum(dbscan.labels_ == -1) /len(dbscan.labels_)
    outlier_percent.append(perc_outliers)
```

**TASK: Create a line plot of the percentage of outlier points versus the epsilon value choice.**

```python
#CODE HERE
sns.lineplot(x=np.linspace(0.001,3,50),y=outlier_percent)
plt.xlabel('Epsilon value')
plt.ylabel('Outlier Percenatage')
```

## DBSCAN with Chosen Epsilon

**Based on the plot created in the previous task, retrain a DBSCAN model with a reasonable epsilon value. Note: For reference, the solutions use eps=2.**

```python
dbscan = DBSCAN(eps=2,min_samples=scaled_X.shape[1])
```

```python
dbscan.fit(scaled_X)
```

**Create a scatterplot of Milk vs Grocery, colored by the discovered labels of the DBSCAN model.**

```python
#CODE HERE
sns.scatterplot(data = df,x='Grocery',y='Milk',hue=dbscan.labels_,palette='Set1')
```

**Create a scatterplot of Milk vs. Detergents Paper colored by the labels.**
