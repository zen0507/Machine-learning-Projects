Now let's move on to PCA.

## Scaling Data

**Use Scikit-Learn to scale the pixel feature dataframe.**

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
scaled_pixels = scaler.fit_transform(pixels)
```

```python
scaled_pixels
```

## PCA

**Perform PCA on the scaled pixel data set with 2 components.**

```python
from sklearn.decomposition import PCA
```

```python
pca_model = PCA(n_components=2)
```

```python
pca_pixels = pca_model.fit_transform(scaled_pixels)
```

**How much variance is explained by 2 principal components.**

```python
np.sum(pca_model.explained_variance_ratio_)
```

**TASK: Create a scatterplot of the digits in the 2 dimensional PCA space, color/label based on the original number_label column in the original dataset.**

```python
pca_pixels[:,0]
```

```python
plt.figure(figsize=(10,6),dpi=150)
sns.scatterplot(x = pca_pixels[:,0],y=pca_pixels[:,1],hue = digits['number_label'].values,
               palette='Set1')
plt.legend(loc=(1.05,0))
```

```python
plt.figure(figsize=(10,6),dpi=150)
plt.scatter(x = pca_pixels[:,0],y=pca_pixels[:,1],c = digits['number_label'].values,
               cmap='Set1')
plt.legend(loc=(1.05,0))
```

**Which numbers are the most "distinct"?**

```python
# You should see label #4 as being the most separated group, 
# implying its the most distinct, similar situation for #2, #6 and #9.
```
