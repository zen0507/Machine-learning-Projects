```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
df = pd.read_csv('../DATA/cancer_tumor_data_features.csv')
```

```python
df.head()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
scaled_X = scaler.fit_transform(df)
```

```python
scaled_X
```

```python
from sklearn.decomposition import PCA
```

```python
pca_model = PCA(n_components=2)
```

```python
pca_model.fit(scaled_X)
```

```python
pca_model.transform(scaled_X)
```

```python
pc_results = pca_model.fit_transform(scaled_X)
```

```python
from sklearn.datasets import load_breast_cancer
```

```python
cancer_dictionary = load_breast_cancer()
```

```python
cancer_dictionary['target']
```

```python
# plt.figure(figsize=(8,6),dpi = 150)
plt.scatter(pc_results[:,0],pc_results[:,1],c=cancer_dictionary['target'])
```

```python
pca_model.components_.shape
```

```python
df_cmp = pd.DataFrame(pca_model.components_,index=['PC1','PC2'],columns=df.columns)
```

```python
df_cmp
```

```python
plt.figure(figsize=(20,3),dpi=150)
sns.heatmap(df_cmp,annot=True)
```

```python
np.sum(pca_model.explained_variance_ratio_)
```

```python
pca_30 = PCA(n_components=30)
```

```python
pca_30.fit(scaled_X)
```

```python
np.sum(pca_30.explained_variance_ratio_)
```

```python
explained_var = []
for n in range(1,30):

    pca = PCA(n_components=n)
    pca.fit(scaled_X)

    explained_var.append(np.sum(pca.explained_variance_ratio_))
```

```python
plt.plot(range(1,30),explained_var)
plt.xlabel('Num of Components')
plt.ylabel('Variance Explained')
```
