```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
df = pd.read_csv('../DATA/cancer_tumor_data_features.csv')
```

```python
df.head()
```

```python
sns.heatmap(df) # it is hard to understand anything in this type of data with a lot of features
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
scaled_X = scaler.fit_transform(df)
```

```python
scaled_X.mean(axis = 0)
```

```python
scaled_X
```

```python
scaled_X - scaled_X.mean(axis = 0)
```

```python
covariance_matrix = np.cov(scaled_X,rowvar=False)
```

```python
covariance_matrix
```

```python
eigen_values , eigen_vectors = np.linalg.eig(covariance_matrix)
```

```python
eigen_values
```

```python
eigen_vectors.shape
```

```python
eigen_values.shape
```

```python
#PCA N Feature Space --> N PC SPACE ---> 2
# How Many PC to  choose? N = 2
```

```python
num_components = 2
```

```python
# sorted_key = np.argsort
```

```python
eigen_values
```

```python
# [0,1,2]
sorted_key = np.eargsort(eigen_values)[::-1][:num_components]
```

```python
eigen_values , eigen_vectors  = eigen_values[sorted_key],eigen_vectors[:,sorted_key]
```

```python
principal_components = np.dot(scaled_X,eigen_vectors)
```

```python
principal_components
```

```python
plt.figure(figsize=(8,6),dpi = 150)
plt.scatter(principal_components[:,0],principal_components[:,1])
```
