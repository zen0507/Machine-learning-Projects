```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd 

import seaborn as sns
```

```python
df = pd.read_csv("StudentsPerformance.csv")
```

```python
df.head()
```

```python
plt.figure(figsize=(10,8),dpi=300)
sns.boxplot(data=df,x='reading score',y = 'parental level of education',hue='test preparation course',palette='Set2')
plt.legend(bbox_to_anchor=(1.05,0.5))
```

```python
plt.figure(figsize=(10,8),dpi=200)
sns.violinplot(data=df,x='reading score',y = 'parental level of education',
                bw_method = 0.5) #split=False,inner='stick'
# plt.legend(bbox_to_anchor=(1.25,0.5))
```

```python
plt.figure(figsize=(8,4),dpi=200)
sns.swarmplot(data = df,x='math score',y='gender',size=2,hue='test preparation course',dodge=True)
plt.legend(bbox_to_anchor=(1.35,0.5))
```

```python
sns.boxenplot(x='math score',y='test preparation course',data = df,hue='gender')
```
