```python
import matplotlib.image as mpimg
import matplotlib.pyplot as plt
```

```python
image_as_array = mpimg.imread('../DATA/zen.jpg')
```

```python
image_as_array.shape # (H,W,C)
```

```python
# image_as_array
```

```python
# plt.figure(dpi=200)
plt.imshow(image_as_array)
```

```python
# (H<W<C) --> 2D (H*W,C)
```

```python
(h,w,c) = image_as_array.shape 
```

```python
c
```

```python
image_as_array2d = image_as_array.reshape(h*w,c)
```

```python
len(image_as_array.shape)
```

```python
len(image_as_array2d.shape)
```

```python
from sklearn.cluster import KMeans
```

```python
model = KMeans(n_clusters=9)
```

```python
labels = model.fit_predict(image_as_array2d)
```

```python
rgb_codes = model.cluster_centers_.round(0).astype(int)
```

```python
# rgb_codes
```

```python
rgb_codes[labels]
```

```python
quantized_image = np.reshape(rgb_codes[labels],(h,w,c))
```

```python

plt.imshow(quantized_image)
plt.axis('off')  # optional: hides axes
plt.savefig("C:\\Users\\LENOVO\\OneDrive\\Pictures\\zen.jpg", bbox_inches='tight', pad_inches=0)
plt.show() 
```
