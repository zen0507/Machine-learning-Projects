```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
a = np.linspace(0,10,11)
b = a**4
```

```python
x = np.arange(0,10)
y = x*2
```

```python
fig,axes = plt.subplots(nrows=2,ncols=2,figsize = (0,7),dpi = 200,projection = "3D")
# axes[0][0].plot(x,y)
# axes[1][1].plot(a,b)
# for ax in axes:
#     ax.plot(x,y)

axes[0][0].plot(x,y)


axes[0][1].plot(x,y)

axes[1][0].plot(x,y)
axes[1][0].set_ylabel('Y Label 1,0')
axes[1][0].set_xlim(2,6)

axes[1][1].plot(a,b)
axes[1][1].set_title('TITLE 1,1')
fig.set_figwidth(10)
fig.suptitle("Figure level") #,fontsize = 16
# plt.subplots_adjust(wspace=1,hspace=0.5)
plt.tight_layout()

fig.savefig("C:\\Users\\LENOVO\\OneDrive\\Pictures\\new_subplots.png",bbox_inches = 'tight')
```
