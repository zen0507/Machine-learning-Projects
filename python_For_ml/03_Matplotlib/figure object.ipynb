## Understanding the figure object

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# 422 *  288 figure with no set of axis - fig = plt.figure()
# to add axis into this we use axes = fig.add_axes([0,0,1,1]) - 
# lower left corner of axes(it ranges from 0 to 1 and 0.5 is middle , this helps to place axes in the figure if it is greater than one axes will be
# outside of the figure) - (x,y) - this rep the first 0,0 in the parameter of fig.add_axes
# for next 2 parameters it rep weight and height and 1,1 means width and height is at max by 0.5 it means half
# this helps to add multiple axes as  well as move and resize the axes
```

```python
#DATA
```

```python
a = np.linspace(0,10,11)
```

```python
b = a**4
```

```python
a
```

```python
b
```

```python
x = np.arange(0,10)
```

```python
y = x*2
```

```python
 x
```

```python
y
```

```python
fig = plt.figure()

# LARGE AXES
axes1 = fig.add_axes([0,0,1,1])
axes1.plot(a,b);

axes1.set_xlim(0,8)
axes1.set_ylim(0,8000)
axes1.set_xlabel('A')
axes1.set_ylabel('B')
axes1.set_title('Power Of 4')


# SMALL AXES
axes2 = fig.add_axes([1,1,2,2])
axes2.plot(x,y);

axes2.set_xlim(0,2)
axes2.set_ylim(0,50)
axes2.set_xlabel('A')
axes2.set_ylabel('B')
axes2.set_title('Zoomed In')
```
