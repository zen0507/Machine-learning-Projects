```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("Advertising.csv")
```

```python
X = df.drop('sales',axis = 1)
```

```python
y = df['sales']
```

```python
from sklearn.preprocessing import PolynomialFeatures
```

```python
polynomial_convertor = PolynomialFeatures(degree=3,include_bias=False)
```

```python
poly_features = polynomial_convertor.fit_transform(X)
```

```python
poly_features.shape
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(poly_features, y, test_size=0.3, random_state=42)
```

```python
X_train.shape
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
scaler.fit(X_train)
```

```python
X_train = scaler.transform(X_train)
```

```python
X_test = scaler.transform(X_test)
```

```python
X_train[0]
```

```python
# poly_features[0]
```

```python
X_test[0]
```

## Ridge Regression / L1 Rgularization


```python
from sklearn.linear_model import Ridge
```

```python
ridge_model = Ridge(alpha=10)
```

```python
ridge_model.fit(X_train,y_train)
```

```python
test_predictions = ridge_model.predict(X_test)
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error
```

```python
MAE = mean_absolute_error(y_test,test_predictions)
```

```python
MAE
```

```python
RMSE = np.sqrt(mean_squared_error(y_test,test_predictions))
```

```python
print(RMSE)
```

```python
from sklearn.linear_model import RidgeCV
```

```python
ridge_cv_model = RidgeCV(alphas=(0.1, 1.0, 10.0),scoring='neg_mean_absolute_error')
```

```python
ridge_cv_model.fit(X_train,y_train)
```

```python
ridge_cv_model.alpha_
```

```python
#from sklearn.metrics import get_scorer
```

```python
#get_scorer_names()
```

```python
test_predictions = ridge_cv_model.predict(X_test)
```

```python
MAE = mean_absolute_error(y_test,test_predictions)
```

```python
RMSE = np.sqrt(mean_squared_error(y_test,test_predictions))
```

```python
MAE
```

```python
RMSE
```

```python
ridge_cv_model.coef_
```

```python
ridge_cv_model.best_score_
```

## LASSO -least absolute shrinkage and selection operator

```python
from sklearn.linear_model import LassoCV
```

```python
lasso_cv_model = LassoCV(eps=0.001,n_alphas = 100,cv = 5, max_iter = 100000) # max_iter = 100000 or eps = 0.001
```

```python
lasso_cv_model.fit(X_train,y_train)
```

```python
lasso_cv_model.alpha_
```

```python
test_predictions = lasso_cv_model.predict(X_test)
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error
```

```python
MAE = mean_absolute_error(y_test,test_predictions)
```

```python
RMSE = np.sqrt(mean_squared_error(y_test,test_predictions))
```

```python
MAE
```

```python
RMSE
```

```python
lasso_cv_model.coef_ # by considering this we can easy to interpret
```

## L1 and L2 Regularization (Elastic Net)

```python
from sklearn.linear_model import ElasticNetCV
```

```python
elastic_model = ElasticNetCV(l1_ratio=[.1, .5, .7, .9, .95, .99, 1],eps=0.001,n_alphas=100,max_iter=1000000)
```

```python
elastic_model.fit(X_train,y_train)
```

```python
elastic_model.l1_ratio_
```

```python
elastic_model.alpha_
```

```python
lasso_cv_model.alpha_
```

```python
test_predictions = elastic_model.predict(X_test)
```

```python
MAE = mean_absolute_error(y_test,test_predictions)
```

```python
RMSE = np.sqrt(mean_squared_error(y_test,test_predictions))
```
